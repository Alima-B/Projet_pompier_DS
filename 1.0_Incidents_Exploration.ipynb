```python
## Data incidents depuis 2018
Incidents_2018 = pd.read_csv("Data/LFB Incident data from 2018 onwards.csv") 
```

```python
Incidents_2018.head()
```

```python
# Modification du type de certaines variables
# CalYear(int) passe en object
# HourOfCall(int) passe en object
# UPRN et USRN(int) passe en object
# Easting_m et Northing_m (float) passent en obj
# Latitutude et Longitude (float) passent en obj

dico = {'CalYear' : 'object',
       'UPRN' : 'object',
       'USRN' : 'object'}

Incidents_2018 = Incidents_2018.astype(dico)
# TimeOfCall et DateOfCall (obj) passe en datetime
Incidents_2018['DateOfCall'] = pd.to_datetime(Incidents_2018['DateOfCall'],format = "%d-%b-%y")
Incidents_2018['TimeOfCall'] = pd.to_datetime(Incidents_2018['TimeOfCall'],format = "%H:%M:%S")
```

```python
Incidents_2018.info()
```

```python
Incidents_2018.head()
```

```python
# check si meme type de data dans les deux tab
list(Incidents_2009.dtypes) == list(Incidents_2018.dtypes)
```

```python
# Meme nom de quartier à partir de 2018
list(Incidents_2018.IncGeo_WardName) == list(Incidents_2018.IncGeo_WardNameNew)
```

```python
list(Incidents_2009.IncGeo_WardName.unique()) == list(Incidents_2009.IncGeo_WardNameNew.unique())
# Pour les incidents de la tab 2009 à 2017 le nom des quartiers n'est pas identique
```

```python
# ligne ou IncGeo_WardName et IncGeo_WardNameNew ne sont pas pareils. 
#Ajout d'un point après St (St. Mary's et St. Peter's)
Incidents_2009[(Incidents_2009.IncGeo_WardName == Incidents_2009.IncGeo_WardNameNew) == False]
```

```python
# CONCAT DES DEUX ET SAUVEGARDE LE DATAFRAME
Incidents = pd.concat([Incidents_2009,Incidents_2018], axis = 0, ignore_index = True)
```

```python
Incidents.head()
```

```python
Incidents.info()
```

```python
# Type d'incident
print(Incidents.IncidentGroup.value_counts(),'\n')
# Detail incident
print(Incidents.StopCodeDescription.value_counts())
```

```python
# Type de service special 
# RTC road traffic collision ?
print('nb variables: 2009:',len(Incidents_2009.SpecialServiceType.value_counts())," 2018:",len(Incidents_2018.SpecialServiceType.value_counts()))
#Meme nombre de variables entre les deux Dataframe 
print("\n",Incidents.SpecialServiceType.value_counts())
```

```python
print('nb variables: 2009:',len(Incidents_2009.IncidentStationGround.value_counts())," 2018:",len(Incidents_2018.IncidentStationGround.value_counts()))
print("\n",Incidents.IncidentStationGround.unique())
print("nb variables total",len(Incidents.IncidentStationGround.unique()))

#Nb de variables diff entre les deux tab
```

```python
set(Incidents_2018.IncidentStationGround.unique()).difference(set(Incidents_2009.IncidentStationGround.unique()))
# Ajout de Beyond Home Counties à partir de 2018. Les lieux limitrophes au grand Londres (Essex, Surrey etc..)
```

```python
# Codes postal
print('nb variables: 2009:',len(Incidents_2009.Postcode_district.unique())," 2018:",len(Incidents_2018.Postcode_district.unique()))
print("\n",Incidents.Postcode_district.unique())
print("\n nb variables total:",len(Incidents.Postcode_district.unique()))

# Nombre de codes postal différents
```

```python
# Postcode de 2018 non présent en 2009
print("Postcode de 2018 non présent en 2009",set(Incidents_2018.Postcode_district.unique()).difference(set(Incidents_2009.Postcode_district.unique())))
# Postcode de 2009 non présent en 2018
print("Postcode de 2009 non présent en 2018",set(Incidents_2009.Postcode_district.unique()).difference(set(Incidents_2018.Postcode_district.unique())))
```

```python
# postcodes de la tab 2018 absents dans 2009
pstcode_2018 = set(Incidents_2018.Postcode_district.unique()).difference(set(Incidents_2009.Postcode_district.unique()))
# postcodes de la tab 2009 absents dans 2018
pstcode_2009 = set(Incidents_2009.Postcode_district.unique()).difference(set(Incidents_2018.Postcode_district.unique()))
```

```python
# Arrondissements
print("quartier de 2018",Incidents["IncGeo_BoroughName"][Incidents["Postcode_district"].isin(list(pstcode_2018))].unique())
print("quartier de 2009",Incidents["IncGeo_BoroughName"][Incidents["Postcode_district"].isin(list(pstcode_2009))].unique())
      
# quartiers
print("quartier de 2018",Incidents["IncGeo_WardNameNew"][Incidents["Postcode_district"].isin(list(pstcode_2018))].unique())
print("quartier de 2009",Incidents["IncGeo_WardNameNew"][Incidents["Postcode_district"].isin(list(pstcode_2009))].unique())
      
```
