```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib as plt
import seaborn as sns
```

```python
Mob_2009 = pd.read_csv("Data/LFB_Mobilisation_data_from_January_2009 - 2014.csv") 
```

```python
Mob_2009.head()
```

```python
Mob_2009.info()
```

```python
# Modification du type de certaines variables
# IncidentNumber passe en object
# CalYear passe en object
# ResourceMobilisationId passe en object
# PumpOrder en object
# DelayCodeId en object
dico = {'IncidentNumber' : 'object',
       'CalYear' : 'object',
       'ResourceMobilisationId' : 'object',
       'PumpOrder' : 'object',
       'DelayCodeId' : 'object'}

Mob_2009 = Mob_2009.astype(dico)
# DateAndTimeMobilised, DateAndTimeMobile et DateAndTimeArrived en Datetime
# DateAndTimeLeft, DateAndTimeReturned en Datetime
# Format de Date : Jour/Mois/Année  donc paramètre format = "%d/%m/%Y %H:%M"
Mob_2009['DateAndTimeMobilised'] = pd.to_datetime(Mob_2009['DateAndTimeMobilised'],format = "%d/%m/%Y %H:%M")
Mob_2009['DateAndTimeMobile'] = pd.to_datetime(Mob_2009['DateAndTimeMobile'],format = "%d/%m/%Y %H:%M")
Mob_2009['DateAndTimeArrived'] = pd.to_datetime(Mob_2009['DateAndTimeArrived'],format = "%d/%m/%Y %H:%M")
Mob_2009['DateAndTimeLeft'] = pd.to_datetime(Mob_2009['DateAndTimeLeft'],format = "%d/%m/%Y %H:%M")
Mob_2009['DateAndTimeReturned'] = pd.to_datetime(Mob_2009['DateAndTimeReturned'],format = "%d/%m/%Y %H:%M")
```
