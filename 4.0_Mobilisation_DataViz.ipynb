# visualisation des donn√©es pour le dataFrame Mobilisation

```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# DataFrame de toute les ann√©es
Mobilisation =pd.read_csv("Data/Mobilisation_2009_2023.csv")

pd.set_option('max_colwidth',50)
pd.set_option('display.max_columns',50)
Mobilisation.head()
```

```python
dico = {'IncidentNumber' : 'object',
       'CalYear' : 'object',
       'ResourceMobilisationId' : 'object',
       'PumpOrder' : 'object',
       'DelayCodeId' : 'object'}

Mobilisation = Mobilisation.astype(dico)

Mobilisation['DateAndTimeMobilised'] = pd.to_datetime(Mobilisation['DateAndTimeMobilised'])
Mobilisation['DateAndTimeMobile'] = pd.to_datetime(Mobilisation['DateAndTimeMobile'])
Mobilisation['DateAndTimeArrived'] = pd.to_datetime(Mobilisation['DateAndTimeArrived'])
Mobilisation['DateAndTimeLeft'] = pd.to_datetime(Mobilisation['DateAndTimeLeft'])
Mobilisation['DateAndTimeReturned'] = pd.to_datetime(Mobilisation['DateAndTimeReturned'])
```
